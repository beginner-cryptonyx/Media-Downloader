```python
import subprocess
import json
```

```python
link = "https://youtu.be/x-wf8Yb5ZCM?si=u1ZWtQc85v1WCFH7"
```

```python
formats = subprocess.run(["yt-dlp", "-j", link], capture_output=True, text=True)
formats = json.loads(formats.stdout)
```

```python
print(type(formats))
print(formats.keys())

video_details = {
    'id': formats['id'],
    'title': formats['title'],
    'duration': formats['duration'],
    'views': formats['view_count'],
    # 'id': formats['id'],
    # 'id': formats['id'],
    # 'id': formats['id'],
}
# for format in formats["formats"]:
# #     print(format)
# # print(formats[0])
#     print(format)
print(video_details)
```

```python
print(type(formats["formats"]))
audio_info = {}
video_info = {}
for format in formats["formats"]:
    if format['format_note'] != 'storyboard':
        if format["vcodec"] == 'none':
            audio_info[format["format_id"]] = {
                "note": format["format_note"],
                "file_ext": format['ext'],
                "size": format["filesize_approx"],
                "avg_bitrate": format["abr"]
            }
        else:
            video_info[format["format_id"]] = {
                "note": format["format_note"],
                "file_ext": format['ext'],
                "size": format["filesize_approx"],
                'fps': format['fps']
            }
    # format.pop()
print(audio_info)
print(video_info)
```

```python
print(audio_info)
sorted_audio = dict(sorted(audio_info.items(), key=lambda item: item[1]["avg_bitrate"], reverse=True))
print(sorted_audio)
```
